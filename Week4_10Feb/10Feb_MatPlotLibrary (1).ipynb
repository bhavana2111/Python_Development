```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
array = np.linspace(0,10,100)
figure= plt.figure()
plt.plot(array,np.sin(array),c='red',ls='-')
plt.plot(array,np.cos(array),c='green',ls='--')
plt.plot(array,np.tan(array),c='blue',ls=':')
```

```python
array = np.linspace(0,10,100)
figure = plt.figure()
plt.subplot(3,1,1)
plt.plot(array,np.sin(array), c='red', ls = '-')
plt.subplot(3,1,2)
plt.plot(array,np.cos(array),c='green',ls = '--')
plt.subplot(3,1,3)
plt.plot(array,np.tan(array),c='blue',ls=':')
```

```python
array = np.linspace(0,10,10)
plt.plot(array, array*array, c='red',ls='-',ms=2,label='Square')
plt.plot(array, array*array*array, c='green',ls='-',ms=2,label='Cube')
plt.plot(array, array*array*array*array, c='blue',ls='-',ms=2,label= "Quaduple")
plt.xlabel("Number")
plt.ylabel("Arithmetic operation result")
plt.title("Graphical Representation")
plt.legend()
plt.show()
```

```python
figure = plt.figure()
# Create data to plot
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]


axis1 = figure.add_axes([0.05,.7,.5,.5])
axis1.set_label("Top Left")
axis1.plot(x,y,'r')

axis2 = figure.add_axes([0.75,.7,.5,.5])
axis2.set_label("Top Right")
axis2.plot(x,y,'g')

axis3 = figure.add_axes([0.05,.05,.5,.5])
axis3.set_label("Bottom Left")
axis3.plot(x,y,'b')

axis4 = figure.add_axes([0.75,.05,.5,.5])
axis4.set_label("Bottom Right")
axis4.plot(x,y,'y')
```
