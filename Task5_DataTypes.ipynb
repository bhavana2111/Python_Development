```python
print("INBUILT STRING FUNCTIONS")
myString = "WELCOME TO THE WORLD OF PYTHON"
print("CAPITALIZE FUNCTION")
print(myString.capitalize())

print("\nCOUNT FUNCTION")
print("Count of alphabet \'w' ",myString.count('w'))
print("Count of alphabet \'t' ",myString.count('t'))
print("\nENDS WITH FUNCTION")
print("Does the string end with th ? ",myString.endswith("th"))
print("Does the string end with on ? ",myString.endswith("on"))
print("\nEXPAND TABS FUNCTION")
myString = "WELCOME\t TO\t THE WORLD OF PYTHON"
print("Expand Tabs = ",myString.expandtabs(20))

print("\nFIND FUNCTION")
myString = "WELCOME TO THE WORLD OF PYTHON"
print("Is ON present in string",myString.find("ON"))
print("\nINDEX FUNCTION")
myString = "WELCOME TO THE WORLD OF PYTHON"
print("Is ON present in string",myString.index("ON"))


print("\nALPHA NUMERIC")
print(myString,"is alpha numeric ? ",myString.isalnum())
myString2 = "12 WELCOME TO THE FIRST WORLD OF PYTHON123"
print(myString2,"is alpha numeric ? ",myString2.isalnum())

print("\nIS DIGIT FUNCTION")
digit = '10'
print(digit, " is a digit ? ",digit.isdigit())

print("\nIS DECIMAL FUNCTION")
decimal = '11'
print(decimal, " is a decimal ? ",decimal.isdecimal())
decimal = '11.45'
print(decimal, " is a decimal ? ",decimal.isdecimal())

print("INDENTIFIER FUNCTION\n")
str = "30Python"
print("Is Identifier ",str,str.isidentifier())
str = "Python"
print("Is Identifier ",str,str.isidentifier())

print("LOWER FUNCTION\n")
str = "30Python"
print("Lower of",str,str.lower())
print("Is {} lower? {}".format(str,str.islower()))
str = "Python"
print("UPPER  ",str,str.upper())
print("Is {} lower? {}".format(str,str.isupper()))


print("\nisNumeric function")
num = '20'
print("{} is numeric = {}".format(num, num.isnumeric()))


print("\n join function")
teams = ["Chennai", "Hyd", "Bangalore", "Orissa"]
print("#".join(teams))


print("\n strip function")
str = "Python learning and practicing"
print(str.strip('l'))
```
