```python
# Print Statement
num1 = 10
num2 = 20
str1 = "Python"
print(num1)
print(num1,num2)
print(num1,num2, str1)
print("Sum of ",num1," and ",num2, " is equal to ",(num1 + num2))


# Print Result with String
name = "Python"
age = "30"
type = "programming language"
print(name," is a ",age,"  year old ",type)

# "format" method with print
name = "Python"
age = "30"
type = "programming language"
print("{} is a {} year old {}".format(name,age,type))

num1 = 20
num2 = 30
num3 = 40
print("Averages of numbers {}, {} and {} is equal to {}".format(num1,num2,num3,((num1+num2+num3)/3)))


# 'f' string method

name = "Python"
age = "30"
type = "programming language"
print(f"{name} is a {age} year old {type}")

num1 = 20
num2 = 30
num3 = 40
print(f'Averages of numbers {num1}, {num2} and {num3} is equal to {num1+num2+num3/3}')


# Combine everything
virat = 100
dhoni = 200
rishabh = 250
print("**********Combine everything**********")
print(virat,",",dhoni,"and",rishabh," are the top 3 scores of the match")
print("{}, {} and {} are the top 3 scores of the match".format(virat,dhoni,rishabh))
print(f"{virat}, {dhoni} and {rishabh} are the top 3 scores of the match")

#"end" statement to join lines
print("Hyderabad",end='--')
print("Chennai",end="--")
print("Bangalore")

#"sep" seperator between words
print("Bangalore to Chennai route is :-")
print("Bangalore","Devanahalli","Anantapur","Kurnool","MehboobNagar","ShadNagar","Shamshabad","Hyderabad",sep="--->")

print("Num1","Num2","Num3","Num4","Num5",sep="#")

print(10,'.',00)
print(10,'.',00,sep="")


# Combine sep and end
print(1,2,3,'.',sep="",end="")
print(4,5,sep="")
```
