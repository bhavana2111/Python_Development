```python
import sys
print(sys.executable)
import ollama
```

```python
def generateInsights(dataFrame):
    prompt = f'Generate Summary of Dataset and provide me the Summary.\n\n{dataFrame}'
    response = ollama.chat(model = 'mistral', messages = [{'role':'user','content':prompt}])
    return response['message']['content']
    
    
summary = titanic.describe().to_string()
insights = generateInsights(summary)
print("\n AI Generated Insights = ",insights)
```

```python
import gradio as gr
def eda_analysis(file):
    df = pd.read_csv(file)
    summary = df.describe().to_string()
    insights = generateInsights(summary)
    return insights
    
gradio = gr.Interface(fn = eda_analysis, inputs = "file", outputs='text')
gradio.launch()
```
